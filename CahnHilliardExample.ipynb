```python
%matplotlib notebook
from matplotlib import pyplot
import matplotlib.animation as animation
import numpy as np

# if you import CahnHilliard, you would access the
# CH class as `CahnHilliard.CH`. Now `CH` has the
# same effect
from CahnHilliard import CH
```

```python
myCH = CH(100, 100, 0.1, 0.0, 5.0, 0.0002)
myCH.initRandom()
t = myCH.t0
```

```python
fig = pyplot.figure(figsize=(6,6))
im = pyplot.imshow(myCH.C, animated=True, vmin=0, vmax=1, origin="lower")

def updatefig(*args):
    global myCH
    for i in range(100):
        myCH.nextC(myCH.C)
    im.set_array(myCH.C)
    return im,

ani = animation.FuncAnimation(fig, updatefig, interval=10, blit=True)
pyplot.show()
```
